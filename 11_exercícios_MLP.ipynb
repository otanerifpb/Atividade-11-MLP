# Exercício 1 - Problema do XOR

```python
## Exercício 1 - Problema do XOR
import numpy as np

x = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 0])
```

```python
# 1- Importar o modelo MLP do sklearn
from sklearn.neural_network import MLPClassifier

# 2- Instanciar o modelo escolhendo uma topologia para a rede, função de ativação e número de épocas de execução até 100% de acerto
model = MLPClassifier(hidden_layer_sizes=(4,2), activation='tanh', max_iter=10000)

#x
```

```python
# 3- Treinar o modelo com os dados de treinamento]
treino = model.fit(x, y)
```

```python
from sklearn import metrics

# 4- Fazer o predict com os dados de treinamento
predictions = treino.predict(x)
acc = metrics.accuracy_score(predictions, y)
show = "{:.2f}".format(acc*100)

# 5- Comparar o resultado do predict com o vetor y
print("Acurácia: {}%".format(show))
print(list(predictions))
print(list(y))
```

# Exercício 2 - Dataset Parkinson

```python
## Exercício 2 - Dataset Parkinsons
import pandas as pd

# 1- Carregar a base de dados Parkinsons
url = "https://raw.githubusercontent.com/tmoura/machinelearning/master/parkinsons.data"

df1 = pd.read_csv(url, header=None)
columns = len(df1.columns)

df1
```

```python
# 2 - Separar o dataset em X (matriz de features) e y (coluna target)

y = df1[0]
X = df1.drop(0, axis=1)

print(y)
X
```

```python
# 3 - Normalizar todas as colunas (normalizar é deixar todos os valores das colunas entre 0 e 1)
from sklearn.preprocessing import MinMaxScaler
#import numpy as np

X[df1.columns[1:]] = MinMaxScaler().fit_transform(df1[df1.columns[1:]])

X
```

```python
# 4- Gerar as bases de treinamento e teste
from sklearn.model_selection import train_test_split
from sklearn import metrics

res1 = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=None, stratify = y)
X_train, X_test, y_train, y_test = res1
```

```python
# 5- Importar o modelo MLP do sklearn
from sklearn.neural_network import MLPClassifier

# 6- Instanciar o modelo escolhendo uma topologia para a rede, função de ativação e número de épocas de execução até que obtenha uma taxa de acerto estável
model1 = MLPClassifier(hidden_layer_sizes=(8,4,5), activation='tanh', max_iter=2000)
```

```python
# 7- Treinar o modelo com os dados de treinamento
treino1 = model1.fit(X_train, y_train)

# 8- Fazer o predict com os dados de teste
predictions1 = treino1.predict(X_test)
acc1 = metrics.accuracy_score(predictions1, y_test)
show1 = "{:.2f}".format(acc1*100)

# 9- Imprimir o percentual de acerto da base de teste
print("Acurácia: {}%".format(show1))
print(list(predictions1))
print(list(y_test))
```

# Exercício 3 - Dataset Penguins

```python
## Exercício 3 - Dataset Penguins

# 1- Carregar a base de dados Penguins da API do Seaborn
from seaborn import load_dataset

# Carregar base de dados
df2 = load_dataset('penguins')

df2
```

```python
# 2- A base precisará ser tratada: Existem valores nulos e dados categóricos
df2 = df2.dropna()

df2.shape

df2
```

```python
# 3- Normalizar todas as colunas (normalizar é deixar todos os valores das colunas entre 0 e 1) --> Redes Neurais são sensíveis as diferenças escalares de valores das features
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

column_transformer = make_column_transformer((OneHotEncoder(), ['island','sex']), remainder='passthrough')
df2 = column_transformer.fit_transform(df2)
df2 = pd.DataFrame(data = df2)

df2
```

```python
from sklearn.preprocessing import MinMaxScaler

df2[df2.columns[6:]] = MinMaxScaler().fit_transform(df2[df2.columns[6:]])
```

```python
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder

label_encoder = preprocessing.LabelEncoder()
df2[5]= label_encoder.fit_transform(df2[5])

df2
```

```python
# 4- Separar o dataset em X (matriz de features) e y (coluna target)
y = df2[5]

X = df2.drop(5, axis=1)

y
```

```python
# 5- Gerar as bases de treinamento e teste
from sklearn.model_selection import train_test_split
from sklearn import metrics

res2 = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=None, stratify = y)
X_train, X_test, y_train, y_test = res2
```

```python
# 6- Importar o modelo MLP do sklearn
from sklearn.neural_network import MLPClassifier

# 7- Instanciar o modelo escolhendo uma topologia para a rede, função de ativação e número de épocas de execução até que obtenha uma taxa de acerto estável
model2 = MLPClassifier(hidden_layer_sizes=(5,4), activation='tanh', max_iter=2000)
```

```python
# 8- Treinar o modelo com os dados de treinamento
treino2 = model2.fit(X_train, y_train)

# 9- Fazer o predict com os dados de teste
predictions2 = treino2.predict(X_test)
acc2 = metrics.accuracy_score(predictions2, y_test)
show2 = "{:.2f}".format(acc2*100)

# 10- Imprimir o percentual de acerto da base de teste
print("Acurácia: {}%".format(show2))
print(list(predictions2))
print(list(y_test))
```
